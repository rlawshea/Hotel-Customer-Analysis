Now we can group each customer by nationality, distribution used to book the hotel, and age group to find out how much these similar groups spend on average. 

```python
agg_df =pd.DataFrame(df.groupby(['nationality','distributionchannel','age_group'])['total_revenue'].agg('mean')).round(2).sort_values("total_revenue")
agg_df.head(20)
```

```python
agg_df=agg_df.reset_index()
agg_df.head(20)
```

By creating a "new user" according to the country and distribution we can try and predict how much other customers with the same demographics will spend on average. 

```python
# making copy of team column
new = agg_df["distributionchannel"].copy()
 
# concatenating team with name column
# overwriting name column
agg_df["nationality_distribution"]= agg_df["nationality"].str.cat(new, sep ="_")
 
# display
agg_df.head(20)
```

## Results 

After analyzing the data we can see that there are some significant findings that should be acknowledged:

1.	This hotel has a great check-in rate since 99.84% of customers ended up checking in for their booking.
2.	Almost 26% of the hotel’s customers were between the ages of 45-54 years old. 21% of customers were between the ages of 35-44, and 18.6% were between the ages of 55-64. 17% of customers were between the ages of 25-34, and 13.7% were over 65. Whereas, only 4% of customers were between the ages of 18-24.
3.	The top 5 nationalities that drew in the most customers were France, Germany, the United Kingdom, Portugal, and Spain. However, the top 5 countries that spent the most on average are Namibia, San Marino, Malawi, Andorra, and the Democratic Republic of the Congo.
4.	The distribution channel that approximately 83% of consumers booked through was travel agents/operators. 12.9% of customers booked directly. 3.5% booked through corporate, and only 0.8% of customers booked through electronic distribution.
5.	Even though 82.8% of customers booked through a travel agent, customers who booked directly spent about $100 more on average.
6.	For booking channels, 18-24-year-olds spend more than the other groups when booked through direct and travel agent distributions. For corporate booking, you can expect 55-64-year-olds to spend more on average. For electronic distribution, 25-34-year-olds spend the most on average.



## Conclusion

After reviewing these findings, we can conclude that the hotel’s booking-to-checked-in conversion strategy is working. Thus, this area needs little to no changes. Since 26.7% of the hotel’s clientele are 45-54-year-olds, the hotel should implement marketing procedures that retain and increase customers with similar demographics. In particular, consumers from the top 5 countries mentioned. Also, since we know that all age groups spend more when booked directly, the hotel should be trying to implement a way you can get more consumers to book with the hotel directly. Surveys on why a customer decided to book through a travel agent versus directly from the hotel can give insight into what is driving the customer’s decision. Offering to collaborate with travel agent organizations to gain insight, and business partners may also be a solution. 

In addition, 18-24-year-olds tend to spend more, if not approximately the same, as the other age groups. Appealing to their lowest demographic could increase revenue. Lastly, the hotel should figure out the reason for the stay. Theoretically, people who stay for business versus vacation will most likely spend less. Increasing the hotel's vacation appeal could pay off. Either way, after doing this analysis, by knowing the customers’ nationality, distribution, and age group, the hotel can see what others with similar backgrounds have spent and predict what they will spend on average.
